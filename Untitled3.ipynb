```
import numpy as np
import matplotlib.pyplot as plt
```

```
x=np.linspace(1,10,5)
y=x**2
```

```
plt.plot(x,y,'ro')
```

```
plt.subplot(2,3,1)
plt.plot(x,y,'r')

plt.subplot(2,3,2)
plt.plot(x,y,'g')

plt.subplot(2,3,3)
plt.plot(x,y,'b')

plt.subplot(2,3,5)
plt.plot(x,y,"pink")

plt.subplot(2,3,4)
plt.plot(x,y,'brown')

plt.subplot(2,3,6)
plt.plot(x,y,'yellow')

#plt.subplot(2,3)
#plt.plot(x,y)
```

```
fig=plt.figure()
axes1=fig.add_axes([0,0,0.8,0.8])
axes1.plot(x,y)

axes1=fig.add_axes([0.4,0.4,0.3,.3])
axes1.plot(x,y)
```
